```python
import pandas as pd
import matplotlib.pyplot as plt

table_base = pd.read_csv("bestsellers with categories.csv")

df = table_base

df

```

```python
num_rows, num_cols = df.shape
print(f"\nКоличество строк: {num_rows}")
print(f"Количество столбцов: {num_cols}")

print("Про скільки книг зберігає дані датасет? Відповідь: 550")
```

```python
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
df
```

```python
missing_counts = df.isnull().sum()
missing = (missing_counts / len(df)) * 100

print(f"Частка пропусків: {missing}")

print("Чи є в якихось змінних пропуски? Відповідь: ні")
```

```python
unique_genres = df['genre'].unique()
print("Які є унікальні жанри? Відповідь:", unique_genres)

unique_genres = df['genre'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(unique_genres, labels=unique_genres.index, autopct='%1.1f%%', startangle=140)
plt.title('Розподіл книг за жанрами')
plt.axis('equal')
plt.show()
```

```python
df['price'].plot(kind='hist', bins=5, edgecolor='black')
plt.title('Розподіл цін')
plt.xlabel('Ціна')
plt.ylabel('Частота')
plt.show()
```

```python
# Визначте, яка ціна у нас максимальна, мінімальна, середня, медіанна (використовуйте функції max, min, mean, median)

max_price = df['price'].max()

min_price = df['price'].min()

mean_price = df['price'].mean()

median_price = df['price'].median()

print(f"Максимальна ціна? Відповідь:  {max_price}")
print(f"Мінімальна ціна? Відповідь: {min_price}")
print(f"Середня ціна? Відповідь: {mean_price}")
print(f"Медіанна ціна? Відповідь: {median_price}")

plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(x=max_price, color='red', linestyle='--', linewidth=2, label=f'Max price: ${max_price:.2f}')
plt.axvline(x=mean_price, color='green', linestyle='-', linewidth=2, label=f'Mean price: ${mean_price:.2f}')
plt.axvline(x=median_price, color='orange', linestyle='-.', linewidth=2, label=f'Median price: ${median_price:.2f}')
plt.title('Розподіл цін на книги')
plt.xlabel('Ціна, $')
plt.ylabel('Кількість книг')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
max_rating = df['user_rating'].max()
print(f"Який рейтинг у датасеті найвищий? Відповідь: {max_rating}")
count_max_rating = df[df['user_rating'] == max_rating].shape[0]
print(f"Скільки книг мають такий рейтинг? Відповідь: {count_max_rating}")
max_reviews_book = df.loc[df['reviews'].idxmax(), 'name']
print(f"Яка книга має найбільше відгуків? Відповідь: {max_reviews_book}")
top2015 = df[df['year'] == 2015]
expensive_book = top2015.loc[top2015['price'].idxmax(), 'name']
print(f"З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча? Відповідь: {expensive_book}")
count_fiction_top_50_2010 = df[(df['genre'] == 'Fiction') & (df['year'] == 2010)].shape[0]
print(f"Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? Відповідь: {count_fiction_top_50_2010}")
count_rating_4_9_2010_2011 = df[(df['user_rating'] == 4.9) & ((df['year'] == 2010) | (df['year'] == 2011))].shape[0]
print(f"Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь: {count_rating_4_9_2010_2011}")
sorted_books_2015_under_8 = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values(by='price')
print(f"Яка книга остання у відсортованому списку? Відповідь: {sorted_books_2015_under_8['name'].iloc[-1]}")

```

```python
grouped= df.groupby('genre')
aggregated_data = grouped.agg({'price': ['min', 'max']})

print("Максимальна ціна для жанру Fiction Відповідь: 82\n"
      "Мінімальна ціна для жанру Fiction Відповідь: 0\n"
      "Максимальна ціна для жанру Non Fiction Відповідь: 105\n"
      "Мінімальна ціна для жанру Non Fiction Відповідь: 0")
aggregated_data
```

```python
grouped_auth= df.groupby('author').agg({'name': 'count'}).reset_index()
author_book_count = grouped_auth.rename(columns={'name': 'book_count'})

print(f"Відповідь: Якої розмірності вийшла таблиця? Відповідь: {author_book_count.shape[0]} рядків, {author_book_count.shape[1]} стовпців\n"
    f"Відповідь: Який автор має найбільше книг? Відповідь:{author_book_count.loc[author_book_count['book_count'].idxmax(), 'author']} \n"
    f"Відповідь: Скільки книг цього автора? Відповідь:{author_book_count.loc[author_book_count['book_count'].idxmax(), 'book_count']}")
```

```python
author_avg_rating = df.groupby('author').agg({'user_rating': 'mean'}).reset_index()
min_rating_author = author_avg_rating.loc[author_avg_rating['user_rating'].idxmin(), 'author']
min_rating = author_avg_rating['user_rating'].min()
print(f"У якого автора середній рейтинг мінімальний? Відповідь: {min_rating_author}")
print(f"Який у цього автора середній рейтинг? Відповідь: {min_rating}")
```

```python
merged_df = pd.concat([author_book_count.set_index('author'), author_avg_rating.set_index('author')], axis=1)
sorted_df = merged_df.sort_values(by=['book_count', 'user_rating'])
print(f"Відповідь: Який автор перший у списку? Відповідь: {sorted_df.index[0]}")
sorted_df
```
